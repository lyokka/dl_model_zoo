```python
from keras.datasets import boston_housing
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import optimizers
```

```python
# load train, test data
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
# display data shape
print("x_train shape : ", x_train.shape)
print("y_train shape : ", y_train.shape)
```

```python
# display one sample
print("x_train sample : ", x_train[0])
print("y_train sample : ", y_train[0])
```

```python
def K_linear_regression(input_dim, lr=1e-6):
    
    model = Sequential()
    model.add(Dense(1, input_dim = input_dim))
    
    sgd = optimizers.SGD(lr=lr)
    model.compile(optimizer=sgd, loss='mse')
    
    return model
```

```python
# build model
model = K_linear_regression(x_train.shape[1])
# train model
loss_hist = model.fit(x_train, y_train, epochs = 500, batch_size=32)
# evaluate model
model.evaluate(x_test, y_test)
# make prediction
preds = model.predict(x_test)
```
