```python
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from keras import optimizers
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# display sample image
import matplotlib.pyplot as plt
ind = 1
plt.imshow(x_train[ind])
plt.title(y_train[ind])
plt.show()
```

```python
# extract class 9, class 8
x_train_1 = x_train[(y_train == 9).reshape(1,-1)[0]]
x_train_2 = x_train[(y_train == 8).reshape(1,-1)[0]]
x_test_1 = x_test[(y_test == 9).reshape(1,-1)[0]]
x_test_2 = x_test[(y_test == 8).reshape(1,-1)[0]]
```

```python
x_train_1.shape
```

```python
import numpy as np
```

```python
# generate binary classification data

# train
x_train_ = np.concatenate((x_train_1, x_train_2), axis=0).reshape(10000,-1)
y_train_ = np.zeros(10000)
y_train_[:5000] = 1
y_train_ = y_train_.reshape(-1,1)

# test
x_test_ = np.concatenate((x_test_1, x_test_2), axis=0).reshape(2000,-1)
y_test_ = np.zeros(2000)
y_test_[:1000] = 1
y_test_ = y_test_.reshape(-1,1)
```

```python
# build model

model = Sequential()
model.add(Dense(1, activation='sigmoid', input_dim = x_train_.shape[1]))

sgd = optimizers.SGD(lr=1e-5)
model.compile(optimizer=sgd, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# train model
loss_hist = model.fit(x_train_, y_train_, epochs = 500, batch_size=256)
```

```python
# evaluate model
test_eval = model.evaluate(x_test_, y_test_)
print("test loss : ", test_eval[0])
print("test acc : ", test_eval[1])
```

```python
# make prediction
preds = model.predict_classes(x_test_)
```

```python
# display acc history
plt.plot(loss_hist.history['acc'])
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(preds, y_test_)
```
